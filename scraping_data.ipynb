### IMPORT LIBRARY

```python
import pandas as pd 
import requests
```

```python
url = 'https://data.covid19.go.id/public/api/prov.json'
headers = {'User-Agent': 'Chrome'}
```

### MEMULAI REQUEST KE WEBSITE DAN MENGAMBIL DATA

```python

indonesia = requests.get(url, headers= headers).json()
```

### MENAMPILKAN DATA

```python
indonesia
```

```python
aku = pd.DataFrame(indonesia['list_data'])
aku
```
